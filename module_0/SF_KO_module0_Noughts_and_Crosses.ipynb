```python
class color: #Эту часть мы еще не проходили. Нашел в интернете такое решение по форматированию строк
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   ITALIC =  '\033[3m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

def greeting():
    """Приветствие и обозначение базовых правил игры"""
    print(color.RED + color.BOLD + " Добро пожаловать в крестики-нолики! " +color.END)
    print()
    print(color.BOLD+ " Короткое описание правил игры (источник Wikipedia)" +color.END)
    print(""" Игроки по очереди ставят на свободные клетки квадратного поля знаки (один всегда Х, другой всегда 0). 
 Первый ход делает игрок, ставящий крестики.
 Выигрывает первый, выстроивший в ряд определенное количество своих фигуры по вертикали, горизонтали или диагонали.""")
    print()
    print(color.BOLD+ " Особенности этой игры" +color.END)
    print(""" В нашей игре размер поля от 3 до 100 задается игроками в начале игры. Поле квадратное.
 Для победы на поле 3 х 3 или 4 х 4 необходимо построить в ряд 3 одинаковые фигуры. Для полей большего размера 5 одинаковых фигур.""")
    print()
    print(" Формат ввода координат: x y")
    print(" x - номер строки")
    print(" y - номер колонки ")
    print()
    print(" Игрок, играющий крестиками, определяется жребием")
    print()
```

```python
import random

def ask_names_field_size():
    """ Просим ввести имена игроков, выбрать размер игрового поля от 3 до 100 и определяем игрока, делающего первый ход (Х)"""
    name_1, name_2 = input(" Введите имя первого игрока : "), input(" Введите имя второго игрока : ")
    names = [name_1,name_2]
    print()

    while True:
      field_size = input(" Введите размер игрового поля от 3 до 100 : ")

      if not(field_size.isdigit() and 3 <= int(field_size) <= 100):
        print(color.RED+ " Введите целое число от 3 до 100" +color.END) 
      else:
        field_size=int(field_size)
        break
  
    win_size = 3 if field_size <= 4 else 5
    print()
    print(f" Для победы необходимо выстроить {win_size} одинаковые фигуры (Х или 0) по горизонтали, вертикали или диагонали")

    player_1 = random.choice(names) # Случайным образом определяем кто ходит первым
    names.remove(player_1)
    player_2 = names[0]
    print()
    print(f" Первым крестиками ходит {player_1}.")

    return player_1, player_2, field_size, win_size
```

```python
def field_definition(field_size):
    """Создаем поле заданного размера, где 0 строка и столбец неизменяемые координаты"""
    field = [[' ']*(field_size + 1) for i in range(field_size + 1)] # Задаем поле нужного размера
    
    for j in range(field_size):
      field[0][j + 1], field[j + 1][0] = str(j + 1), str(j + 1) # Добавляем координаты в 0 колонку и строку
    
    return field
```

```python
def show_field(field):
    """Отображаем игровое поле"""
    for i, row in enumerate(field):
      row_joined = f" {str(' | '.join(row))} "
      print(" " + row_joined)
      print(" " + "-" * len(row_joined))
```

```python
def ask(player):
    while True:
      cords=input(f" {player}, введите координаты хода:  ").split()
    
      if len(cords) != 2 :
        print(color.RED+ f" Введите через пробел 2 целочисленные координаты внутри игрового поля 1-{field_size}!" +color.END)
        continue
      x, y = cords

      if not(x.isdigit() and 1<=int(x)<=field_size ) or not(y.isdigit() and 1<=int(y)<=field_size):
        print(color.RED + f" Введите через пробел 2 целочисленные координаты внутри игрового поля 1-{field_size}!" +color.END)
        continue
    
      x, y = int(x),int(y)

      if field[x][y]!= " ":
        print(color.RED+ " Эта клетка уже занята, введите другие координаты" +color.END)
        continue

      return x, y
```

```python
def define_win_coords(x,y,win_length,field_size):
    x_min = 1 if x - win_length < 1 else x - win_length
    y_min = 1 if y - win_length < 1 else y - win_length

    x_max = field_size if x + win_length > field_size else x + win_length - 1
    y_max = field_size if y + win_length > field_size else y + win_length - 1
    
    column_win_coords = []
    for pos_x in range(x_min, x_max + 1):
      column_win_coords.append([pos_x,y])

    raw_win_coords = []
    for pos_y in range(y_min, y_max + 1):
      raw_win_coords.append([x,pos_y])
   
    diag1_win_coords = []
    for i in range (win_length):
      if (x - win_length + i) >= 1 and (y - win_length + i) >= 1:
        diag1_win_coords.append([x - win_length + i, y - win_length +i])

    for j in range (win_length):
      if (x + j) <= field_size and (y + j) <= field_size:
        diag1_win_coords.append([x + j, y + j])

    diag2_win_coords = []
    for i in range (win_length,0,-1):
      if (x - i) >= 1 and (y + i) <= field_size:
        diag2_win_coords.append([x - i , y + i])

    for j in range (win_length):
      if (x + j) <= field_size and (y - j) >= 1:
        diag2_win_coords.append([x + j, y - j])

    return column_win_coords, raw_win_coords, diag1_win_coords, diag2_win_coords

```

```python
def check_win(win_coords,field,win_length,player):
    X_win = "".join(["X"]*win_length)
    O_win = "".join(["0"]*win_length)

    symbols_list = [] 
    for cord in win_coords:
      symbols_list.append(field[cord[0]][cord[1]])
      symbols = ''.join(symbols_list)

      if X_win in symbols:
        print(color.GREEN + color.BOLD + color.UNDERLINE + f"Выиграл X {player}!!!" + color.END)
        return True            
      if O_win in symbols:
        print(color.GREEN + color.BOLD + color.UNDERLINE + f"Выиграл 0 {player}!!!" + color.END)
        return True

    return False
```

```python
greeting()
name_1, name_2, field_size, win_length = ask_names_field_size()
field=field_definition(field_size)
count = 0
show_field(field)

while True:
    count += 1

    if count % 2 == 1: 
      print(f" Ходит крестик")
      player = name_1
    else: 
      print(f" Ходит нолик")
      player = name_2
  
    x, y = ask(player)

    if count % 2 == 1:
      field[x][y] = "X"
    else:
      field[x][y] = "0"

    show_field(field)    
    column_win_coords, raw_win_coords, diag1_win_coords, diag2_win_coords = define_win_coords(x,y,win_length,field_size)

    if  check_win(column_win_coords,field,win_length,player):
      break
    
    if  check_win(raw_win_coords,field,win_length, player):
      break

    if  check_win(diag1_win_coords,field,win_length,player):
      break
    
    if  check_win(diag2_win_coords,field,win_length, player):
      break

    if count == field_size**2:
      break
      print("Ничья!")
```
